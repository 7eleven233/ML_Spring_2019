```python
def gen_data(num):
    """
    Function to generate random data
    
    Parameters
    -----------
    num: Integer.  Number of pairs to generate
    
    Returns
    -------
    Tuple (X,Y):
    - X NumPy ndarray, shape (num, 1)
    - Y NumPy ndarray, shape (num, 1)
    """
    v = 2
    t = np.arange(0, num).reshape(-1,1)
    d= (v * t) + np.random.normal(0, np.sqrt(num), (num,1))
    
    return t,d

X, Y = gen_data(50)
```
